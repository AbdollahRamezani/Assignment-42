```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import ListedColormap
```

```python
female_data = pd.read_csv("ANSUR II FEMALE Public.csv")
female_data.head()
```

```python
male_data = pd.read_csv("ANSUR_II_MALE_Public.csv", encoding="latin-1")
male_data.head()
```

```python
data = pd.concat([female_data, male_data])  #به هم چسباندن دو دیتا فریم توسط کتابخانه پانداز
data
```

```python
#Preprocess

data["weightkg"] = data["weightkg"] / 10  # convert to kg
data["stature"] = data["stature"] / 10    # convert to cm
data["Gender"] = data["Gender"].replace(["Female", "Male"], [0, 1])  # convert to binary
```

```python
cmap=['red', 'blue']
plt.scatter(data["stature"], data["weightkg"], c=data["Gender"], alpha=0.1, cmap= ListedColormap(cmap))
plt.title('weight and stature')
plt.xlabel(' stature (cm)')
plt.ylabel('Weight (Kg)')
plt.show()
```

```python
women_stature = data.query("Gender == 0 ")["stature"]
men_stature   = data.query("Gender == 1 ")["stature"]

plt.hist(men_stature ,color= 'blue'  , edgecolor="black"  , alpha = 0.6)
plt.hist( women_stature ,color="red" , edgecolor="black" , alpha = 0.4)
plt.title("women and  men height(cm)")
plt.legend(["men" , "women" ])
plt.xlabel("height(cm)")
plt.ylabel("count")
```

A. Why is the data of men higher than the data of women?
- Because there are more men than women

B. Why is the data of men more right than the data of women?
- Because the average height of men is higher than women

```python
men_stature   = data.query("Gender == 1 ")["stature"]
women_stature = data.query("Gender == 0 ")["stature"]


average_men_stature = men_stature.mean()
average_women_stature = women_stature.mean()

print(f"average_men_stature: {average_men_stature}")
print(f"average_women_stature: {average_women_stature}")
```
